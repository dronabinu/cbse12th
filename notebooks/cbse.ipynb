```python
import pymysql
import pandas as pd
```

```python
db_name = "cbse"
db_host = "mysql-container"
db_username = "root"
db_password = "test1234"

try:
    conn = pymysql.connect(host = db_host,
                           port = int(3306),
                           user = "root",
                           password = db_password,
                           db = db_name)
except e:
    print (e)
if conn:
    print ("connection successful")
else:
    print ("error")
```

```python
cursor= conn.cursor()
  
# execute your query 
cursor.execute("SELECT * FROM students") 
  
# fetch all the matching rows  
result = cursor.fetchall() 
  
# loop through the rows 
for row in result: 
    print(row) 
    print("\n") 
```

```python
def findRoll(roll):
    cursor= conn.cursor()
      
    # execute your query 
    cursor.execute("SELECT * FROM `students` WHERE `roll` = %s", (roll)) 
      
    # fetch all the matching rows  
    result = cursor.fetchall() 
      
    # loop through the rows 
    for row in result: 
        print(row) 
        print("\n") 

findRoll(1)
```

```python
def findName(name):
    cursor= conn.cursor()
      
    # execute your query 
    cursor.execute("SELECT * FROM `students` WHERE `first_name` LIKE %s", (name)) 
      
    # fetch all the matching rows  
    result = cursor.fetchall() 
      
    # loop through the rows 
    for row in result: 
        print(row) 
        print("\n") 

findName("dru")
```

```python
x=int(input("Enter the roll  no:"))
print("You have entered the roll no value as ", x)
findRoll(x)
```

```python
inputname=input("Enter the name:")
print("You have entered the name as " , inputname)
findName(inputname)
```

```python
def findAge(fromAge, toAge):
    cursor= conn.cursor()
      
    # execute your query 
    cursor.execute("SELECT * FROM `students` WHERE `age` BETWEEN %s AND %s", (fromAge, toAge)) 
      
    # fetch all the matching rows  
    result = cursor.fetchall() 
      
    # loop through the rows 
    for row in result: 
        print(row) 
        print("\n") 

findAge(14, 19)
```

```python
fromAge = int(input("Enter the min age:"))
toAge = int(input("Enter the max age:"))
print("Searching between fromAge %s, toAge %s ", fromAge, toAge)


findAge(fromAge, toAge)
```
